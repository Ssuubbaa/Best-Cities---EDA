```python
sns.set(rc={"figure.figsize" :(15,5)})
ax = sns.barplot(data=total,x='city',y='total score')
ax.tick_params(axis='x',rotation=90)
```

```python
sns.set(rc={"figure.figsize" :(15,10)})
ax = sns.barplot(data=business,x='city',y='business score')
ax.tick_params(axis='x',rotation=90)
```

```python
sns.set(rc={"figure.figsize" :(15,10)})
ax = sns.barplot(data=quantity,x='city',y='quatity score')
ax.tick_params(axis='x',rotation=90)
```

```python
sns.set(rc={"figure.figsize" :(15,10)})
ax = sns.scatterplot(data=quality_score,x='city',y='quality score')
ax.tick_params(axis='x',rotation=90)
```

```python
tc = df.groupby(['city'])['total score'].sum().sort_values(ascending=False).nsmallest(10).to_frame()
tc = tc.reset_index()
```

```python
sns.set(rc={"figure.figsize" :(15,10)})
ax = sns.barplot(data=tc,x='city',y='total score')
ax.tick_params(axis='x',rotation=90)
```

```python
df= df.head(10)[['city','quatity score', 'quality score',
                              'business score']].set_index('city')
fig, ax=plt.subplots(figsize=(15,6))
df.plot(kind='bar', stacked=True, color=['wheat','slateblue','magenta'], ax=ax)
ax.set_title('Top 10 Countries for Start-up (3 Metrics)', fontsize=15);
```
